This project explores the correlation between a country's GDP and life expectancy, analyzing trends over time and across nations.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# load data set
df = pd.read_csv('all_data.csv')
df.head()
```

```python
# rename the column
df.rename(columns= {'Life expectancy at birth (years)' : 'LEABY'}, inplace=True)
```

```python
df.head(1)
```

Explore plots through visualization

```python
# GDP
sns.histplot(x=df.GDP, kde=False, bins=30)
plt.xlabel('GDP')
plt.title('GDP Graph')
```

The GDP graph looks like right skewed. This distribution can be described as Power law

```python
df.Year.value_counts()
```

```python
sns.histplot(x=df.LEABY, kde=False, bins=30)
plt.xlabel('Life expectancy at birth (years)')
plt.title('Life expectancy at birth (years) Graph')
```

Life expectancy at birth (years) graph is totally opposite of GDP graph. 'LEABY' graph is left skewed graph

```python
sns.barplot(x='Country', y='LEABY', data=df)
plt.xticks(rotation=30)
```

all the country have above 70 'LEABY' except Zimbabwe

```python
# country vs GDP graph
sns.barplot(y='Country', x='GDP', data=df)
```

for the graph country by GDP America have the heighest GDP than the other counties. Zimbabwe not have any GDP and Chile have very less GDP and  Mexico, Germany, and China have some GDP less than America

```python
sns.lineplot(y = 'GDP', x='Year', data=df, hue='Country')
```

US, China have high GDP growth over the year. Other countries dont have growth like the US, china country but this counties have moderate growth

```python
sns.lineplot(y = 'LEABY', x='Year', data=df, hue='Country')
```

The Zimbabwe have high 'Life expectancy at birth (years)' growth over the year. The other counties also  have increase in  Life expectancy at birth (years) over the year

```python
g = sns.relplot(data = df, x='LEABY', y='GDP', col='Country', hue='Country', kind='scatter', col_wrap=3, height=3, aspect=1, palette = 'colorblind', facet_kws={'sharex':False, 'sharey':False})
g.set_axis_labels("Life expectancy at birth (years)", "GDP in Trillions of U.S. Dollars")
```

Most countries like US, Mexico, Zimbabwe show a linear relationship between GDP and life expectancy.
Chinas trend appears slightly exponential, while Chile's resembles a logarithmic curve.
Overall, there's a positive correlation between GDP and life expectancy across countries.
